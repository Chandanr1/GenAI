```
from re import VERBOSE
prompt_template_name = PromptTemplate(
    input_variables=['cuisine'],
    template=" i want to open a restaurant for {cuisine} food . suggest me few names for this restaurant "
)
name_chain=LLMChain(llm=llm,prompt=prompt_template_name,verbose=True)

item_name=PromptTemplate(
    input_variables=['restaurant_name'],
    template="Suggest some menu items for {restaurant_name}"
)
item_chain=LLMChain(llm=llm,prompt=item_name,verbose=True)
```

```
from langchain.chains import SimpleSequentialChain
chain=SimpleSequentialChain(chains=[name_chain,item_chain])
```

```
import os
os.environ["SERPAPI_API_KEY"]=''
```

```
from langchain.agents import AgentType , initialize_agent , load_tools
from langchain.llms import OpenAI
```

```
llm=OpenAI(temperature=0.9,api_key=os.environ['OPENAI-API-KEY'])
```

```
tools=load_tools(['serpapi'],llm=llm)
agent= initialize_agent(tools,llm=llm,agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,verbose=True)
agent.run("Who won the last mens cricket ODI worldcup")
```

```
!pip install wikipedia
tools=load_tools(['wikipedia','llm-math'],llm=llm)
agent= initialize_agent(tools,llm=llm,agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,verbose=True)
agent.run("in what year was the film saiyaara released ? What is the year raise to the 2 power")
```

```
from langchain.prompts import PromptTemplate
prompt_template_name=PromptTemplate(
    input_variables=['product'],
    template="what is a good name for a company that makes {product}"
)
```

```
from IPython.display import display , Markdown
def display_markdown(text):
  display(Markdown(text))
```

```
from langchain.chains.base import Chain
from langchain.chains import LLMChain
from langchain.memory import ConversationBufferMemory
memory=ConversationBufferMemory()
chain_with_memory=LLMChain(llm=llm,prompt=prompt_template_name,memory=memory)
response=chain_with_memory.run('Motorcycle')
display_markdown(response)
```

```
display_markdown(chain_with_memory.memory.buffer)
```

```
from langchain.chains import ConversationChain
convo=ConversationChain(llm=llm)
display_markdown(convo.prompt.template)
```

```
convo.run("what is the capital of france")
```

```
print(convo.memory.buffer)
```
