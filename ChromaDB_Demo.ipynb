```python
from langchain.document_loaders import DirectoryLoader
from langchain.document_loaders import TextLoader

loader = DirectoryLoader('/content/Data', glob = './*.txt', loader_cls = TextLoader)

documents = loader.load()

print(documents)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_docs(documents, chunk_size = 1000, chunk_overlap = 20):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(documents)
  return docs

docs = split_docs(documents)
print(len(docs))
```

```python
from langchain.embeddings import SentenceTransformerEmbeddings

embeddings = SentenceTransformerEmbeddings(model_name = 'all-MiniLM-L6-v2')
```

```python
from langchain.vectorstores import Chroma

db = Chroma.from_documents(docs, embeddings)
```

```python
db
```

```python
query = 'What are the different kinds of pets people commonly own?'

matching_docs = db.similarity_search(query)
```

```python
matching_docs[0]
```

```python
print(matching_docs[0].page_content)
```

```python
matching_docs = db.similarity_search_with_score(query, k=2)
matching_docs
```

# Persist a chromaDB Instance

```python
persist_dir = 'chroma_db'

vectordb = Chroma.from_documents(
    documents = docs, embedding = embeddings, persist_directory = persist_dir
)
```

```python
vectordb.persist()
```

```python
new_db = Chroma(persist_directory = persist_dir, embedding_function = embeddings)
```

```python
matching_docs = new_db.similarity_search_with_score(query, k=2)
matching_docs
```

```python
from langchain.chat_models import ChatOpenAI

model_name = 'gpt-3.5-turbo'

llm = ChatOpenAI(model_name = model_name)
```

```python
from langchain.chains.question_answering import load_qa_chain

chain = load_qa_chain(llm, chain_type = 'stuff', verbose = True)
```

```python
query = 'What are the emotional benefits of owing a pet?'

matching_docs  = db.similarity_search(query)

# The process of passing revelant correct documents to aid the LLM is called LLM grounding.

answer = chain.run(input_documents = matching_docs, question = query)
answer

```

Retreival QA

```python
from langchain.chains import RetrievalQA

retreival_chain = RetrievalQA.from_chain_type(llm = llm, chain_type = 'stuff', retriever = db.as_retriever())

retreival_chain.run(query)
```
